**Task 2: Control Flow (2 Points)**

Create an array containing the first 10 letters of the alphabet. 
Iterate through the array using a for loop. 
Inside the for loop, use an if statement to print the word "four" instead of the letter "d" and "eight" instead of the letter "i", otherwise print the alphabet letter.

```python
# Create an array containing the first 10 letters of the alphabet. 
array = np.array(['a','b','c','d','e','f','g','h','i','j'])

for alphabet in array:
    if alphabet == 'd':
        print ("four")
    elif alphabet == 'i':
        print ("eight")
    else:
        print (alphabet)
```

**Task 4: Sequence Function (3 Points)**

Create a list containing numbers from 11 to 16, then using a for loop and proper indexing, produce the following output:
12 --> 11

13 --> 12

14 --> 13

15 --> 14

16 --> 15

17 --> 16

```python
# Create a list containing numbers from 11 to 16,
numbers = [11, 12, 13, 14, 15, 16]

# Create a new list with the “shifted” numbers on the left
new_list = [12, 13, 14, 15, 16, 17]

# zip new list and numbers together
for left, right in zip(new_list, numbers):
    print(f"{left} --> {right}")
```

By copying the previous list, create a new variable containing the same numbers from 11 to 16 elements from the previous list. Change an element in one of the variables and make sure the other doesn't change.

**Task 6: Exception Handling (2 Points)**

Create a variable containing a tuple object. Then try to change its first element the way you change an element in a list and see what kind of error you will get. Then try to catch handle the error and instead of the error print "Tuples are immutable :)" 

```python
# Create a variable containing a tuple object. 
tuple_test = (1, 2, 3, 4, 5)

# try to change its first element 
tuple_test[0] = 100
print (tuple_test)
```

```python
try:
    tuple_test = (1, 2, 3, 4, 5)
    tuple_test[0] = 100
    print (tuple_test)

except TypeError:
    print ("Tuples are immutable :)")
```
